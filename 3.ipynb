```python
import matplotlib.pyplot as plt

# Step 1: Generate the dataset
people = ['kiran', 'arun', 'vijay', 'varun']
age = [25, 30, 35, 40]
height = [145, 151, 165, 173]
weight = [45, 55, 65, 75]

# Step 2: Scatter Plot
plt.figure(figsize=(8, 6))
plt.scatter(age, height, color='blue', label='Height')
plt.scatter(age, weight, color='red', label='Weight')
plt.title('Scatter Plot of Age vs. Height and Weight')
plt.xlabel('Age')
plt.ylabel('Height and Weight')
plt.legend()
plt.grid(True)
plt.show()

# Explanation: This scatter plot shows the relationship between age and both height and weight for each person.

# Step 3: Bar Chart
plt.figure(figsize=(8, 6))
bar_width = 0.35
index = range(len(people))
plt.bar(index, height, width=bar_width, color='blue', label='Height')
plt.bar([i + bar_width for i in index], weight, width=bar_width, color='red', label='Weight')
plt.xlabel('People')
plt.ylabel('Height and Weight')
plt.title('Bar Chart of Height and Weight by Person')
plt.xticks([i + bar_width/2 for i in index], people)
plt.legend()
plt.show()

# Explanation: This bar chart compares the height and weight of each person in the dataset.

# Step 4: Histogram
plt.figure(figsize=(8, 6))
plt.hist(height, bins=10, color='blue', alpha=0.7, label='Height')
plt.hist(weight, bins=10, color='red', alpha=0.7, label='Weight')
plt.title('Histogram of Height and Weight')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()

# Explanation: This histogram shows the distribution of height and weight values in the dataset.

```
